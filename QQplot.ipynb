```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

def qqplot_norm(test_params, compare_params, n = 10_000):
    test = np.random.normal(test_params[0], test_params[1], n)
    compare = np.random.normal(compare_params[0], compare_params[1], n)
    
    sns.kdeplot(test, label="empricial dist")
    sns.kdeplot(compare, label="theoretical dist")
    plt.legend();
    # Comparison dist can be specified too here
    sm.qqplot(test, loc = compare_params[0], scale = compare_params[1], line='45')
    plt.show()
    
qqplot_norm([0,5], [0,5])
```

# A shift of distribution

```python
qqplot_norm([5,5], [0,5])
```

## Wider and Narrower Distributions

```python
qqplot_norm([0,2.5], [0,5])
```

```python
qqplot_norm([0,10], [0,5])
```

```python
qqplot_norm([5,10],[0,5])
```

# Comparing different distributions

```python
from scipy.stats import uniform

n = 10_000

test = np.random.normal(0, 5, n)
compare = np.random.uniform(-10, 10, n)

sns.kdeplot(test, label="empricial dist")
sns.distplot(compare, label="theoretical dist")
plt.legend();
# Comparison dist can be specified too here
sm.qqplot(test, dist=uniform, loc=-10,scale=10, line='45')
plt.show()
```

# Das toch niet normaal?

```python
from scipy.stats import lognorm, expon

n = 10_000

test = np.random.lognormal(5, 5, n)
compare = np.random.exponential(5, n)

sns.kdeplot(test, label="empricial dist", clip=(0,100))
sns.kdeplot(compare, label="theoretical dist", clip=(0,100))
plt.legend()
# Comparison dist can be specified too here
sm.qqplot(test, dist=expon, line='45')
plt.show()
```
