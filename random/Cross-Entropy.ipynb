# Plotting Cross-Entropy for different $y$ and $\hat{y}$ values

```python
from math import log
import numpy as np

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

#### $$\log\mathcal{L}\left(\beta\right) = - \frac{1}{N}\sum_{i=1}^N y_i\log\left(\hat{y}_\beta(x_i)\right) + (1 - y_i)\log\left(1 - \hat{y}_\beta(x_i)\right)$$ 

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

Y, YHAT = np.meshgrid(np.linspace(0,1,50), np.linspace(0,1,50))
CE = Y*np.log(YHAT) + (1 - Y)*np.log(1 - YHAT)

ax.plot_wireframe(X=Y, Y=YHAT, Z=CE, cmap=cm.coolwarm)
plt.show()
```

```python
plt.figure(figsize=(12,12))
CS = plt.contour(Y, YHAT, CE, 25)
plt.clabel(CS, inline=1, fontsize=10)
plt.xlabel("actual class", fontsize=18)
plt.ylabel("predicted probability", fontsize=18)
```
